## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df1 = pd.read_csv('data/hist_data.csv')
plt.hist(df1['aデータ'], bins=30, color='orange', alpha=0.5, label='aデータ')
plt.hist(df1['bデータ'], bins=30, color='cyan', alpha=0.5, label='bデータ')
plt.xlabel('温度(℃)')  # x軸ラベル
plt.ylabel('度数')  # y軸ラベル
plt.legend()        # 凡例を表示
plt.title('aデータとbデータの温度に関する度数分布')
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df2 = pd.read_csv('data/scatter_data.csv')
df_class0 = df2[df2["class"] == 0]
x_0 = df_class0["length (cm)"]
y_0 = df_class0["width (cm)"]
plt.scatter(x_0,y_0, s = 20,c = 'red', alpha = 1.0, label = 'Class 0')
df_class1 = df2[df2["class"] == 1]
x_1 = df_class1["length (cm)"]
y_1 = df_class1["width (cm)"]
plt.scatter(x_1,y_1, s = 20,c = 'blue', alpha = 1.0, label = 'Class 1')
df_class2 = df2[df2["class"] == 2]
x_2 = df_class2["length (cm)"]
y_2 = df_class2["width (cm)"]
plt.scatter(x_2,y_2, s = 20,c = 'green', alpha = 1.0, label = 'Class 2')
plt.xlabel("Length (cm)")
plt.ylabel("Width (cm)")
plt.title("クラス0～2のlengthとwidthに関する散布図")
plt.legend()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df3 = pd.read_csv('data/line_data.csv')
x = df3["期間"]
y_1 = df3["件数"]
y_2 = df3["費用"]
fig_1 = plt.figure()
ax_1 = fig_1.add_subplot(1,1,1)
ax_1.plot(x, y_1, linestyle = '-', c = "steelblue")
plt.scatter(x,y_1, s = 20,c = 'steelblue')
plt.xticks(rotation=30)
ax_1.grid()
ax_1.set_xlabel('期間')
ax_1.set_ylabel('件数')
plt.title('期間ごとの件数の推移')
plt.show()
ax_2 = plt.figure()
fig_2, ax_2 = plt.subplots(1, 1) 
ax_2.bar(x, y_2, color = "steelblue")
plt.xticks(rotation=30)
ax_2.grid()
ax_2.set_xlabel('期間')
ax_2.set_ylabel('費用(円)')
plt.title('期間ごとの費用の推移')
plt.show()
```
